```python
import pandas as pd
from sklearn import preprocessing
```

```python
# Import raw data
data = pd.read_csv("../de_id_data.csv")
data
```

### Split out data for use case 1

```python
# Split out data for use case 1
useCase1 = data.loc[data["use_case"]  == 1]
useCase1
```

```python
# Export data before transformation
useCase1.to_csv("../useCase1_raw.csv", index=False)
```

### Transform categorical data to numeric data

```python
useCase1.info()
```

```python
# Encode the ethnicity column

# Create a label encoder object
le = preprocessing.LabelEncoder()

# Fit the label encoder
label = le.fit_transform(useCase1["ethnicity"])

# Add the encoded column to the data
useCase1["ethnicity_trans"] = label

useCase1
```

```python
# Encode the tfc_name column

# Create a label encoder object
le = preprocessing.LabelEncoder()

# Fit the label encoder
label = le.fit_transform(useCase1["tfc_name"])

# Add the encoded column to the data
useCase1["tfc_name_trans"] = label
```

```python
# Encode the patient_on_multiple_pathways column

# Create a label encoder object
le = preprocessing.LabelEncoder()

# Fit the label encoder
label = le.fit_transform(useCase1["patient_on_multiple_pathways"])

# Add the encoded column to the data
useCase1["patient_on_multiple_pathways_trans"] = label
```

```python
# Encode the appt_month column

month_mapping ={
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

# Map the month names to numbers
useCase1["appt_month_trans"] = useCase1["appt_month"].map(month_mapping)

useCase1[['appt_month', 'appt_month_trans']]
```

```python
# Encode the hospital_service_display column

# Create a label encoder object
le = preprocessing.LabelEncoder()

# Fit the label encoder
label = le.fit_transform(useCase1["hospital_service_display"])

# Add the encoded column to the data
useCase1["hospital_service_display_trans"] = label
```

```python
# Encode the service_mapping column

# Create a label encoder object
le = preprocessing.LabelEncoder()

# Fit the label encoder
label = le.fit_transform(useCase1["service_mapping"])

# Add the encoded column to the data
useCase1["service_mapping_trans"] = label
```

```python
# Encode the reason_display column

# Create a label encoder object
le = preprocessing.LabelEncoder()

# Fit the label encoder
label = le.fit_transform(useCase1["reason_display"])

# Add the encoded column to the data
useCase1["reason_display_trans"] = label
```

```python
# Encode the appointment_type column

# Create a label encoder object
le = preprocessing.LabelEncoder()

# Fit the label encoder
label = le.fit_transform(useCase1["appointment_type"])

# Add the encoded column to the data
useCase1["appointment_type_trans"] = label
```

```python
# Encode the did_not_attend_risk_group column

risk_mapping ={
    "Low Risk": 1,
    "Moderate Risk": 2,
    "High Risk": 3,
    "Very High Risk": 4,
}

# Map the month names to numbers
useCase1["did_not_attend_risk_group_trans"] = useCase1["did_not_attend_risk_group"].map(risk_mapping)

useCase1[['did_not_attend_risk_group', 'did_not_attend_risk_group_trans']]
```

### Export the transformed use case 1 data

```python
# Dropped all transformed columns
useCase1_trans = useCase1.drop(["ethnicity",
"tfc_name",
"patient_on_multiple_pathways",
"appt_month",
"hospital_service_display",
"service_mapping",
"reason_display",
"appointment_type",
"did_not_attend_risk_group"], axis =1)

# Export data for use case 1 to CSV
useCase1_trans.to_csv("../useCase1_trans.csv", index=False)
```
